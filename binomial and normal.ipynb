# Binomial distribution

```python
# binomial distribution is a discrete

# it describes the out put of binary scenarios, e.g.toss of a coin

# It has three parameters:

# n - number of trails 
```

```python
# Example 
# Given 10 trails for a coin toss generate 10 data points

from numpy import random

x= random.binomial (n=10, p=0.5, size=10)
print (x)
```

```python
# visualization of binomial distribution 
from numpy import random
import matplotlib.pyplot as plt

import seaborn as sns
sns.distplot(random.binomial(n=10,p=0.5,size=1000),hist=True, kde=False)

plt.show()
```

```python
# Example 
# generate a random normal distribution of 2x3 with mean at 1 and standard deviation at 2:

from numpy import random

x=random.normal(loc=1,scale=2,size=(2,3))

print(x)
```

```python
# visualization of normal distribution 

from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.normal(size=1000), hist=False )

plt.show()
```

```python
# Difference between normal and binomial distribution
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns  

sns.disptlot(random.normal(loc=50,scales=5,size=1000), label='normal')
sns.distplot(random.binomial(n=100,p=0.5,size=1000), label='binomial')

plt.show()
```

```python
# difference between normal and Binomial Distribution
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.normal(loc=50, scale=5, size=1000), hist=False, label='normal')
sns.distplot(random.binomial(n=50, p=0.5, size=1000),hist=False, label='binomial')
plt.legend()
plt.show()
```
